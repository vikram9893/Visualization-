```python
Ans 2 Scatter Plot: Scatter plots are used to display the relationship between two variables. 
They represent individual data points as markers on a two-dimensional plane. 
Scatter plots are created using the scatter function in Matplotlib. 
They are helpful in identifying patterns, clusters, or correlations between variables.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y)

# Adding title, xlabel, and ylabel
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Displaying the plot
plt.show()
```

```python
Ans 3 The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.
It allows you to display multiple plots or axes in a single figure window, arranged in a grid-like structure.

The subplot() function takes three arguments: nrows, ncols, and index. 
The nrows and ncols parameters specify the number of rows and columns in the subplot grid,
while the index parameter specifies the position of the current subplot within the grid.

Here's an example code that demonstrates how to use the subplot() function to create four line plots:
import numpy as np
import matplotlib.pyplot as plt

# Line 1 data
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2 data
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3 data
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4 data
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# Plotting the first subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

# Plotting the second subplot
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

# Plotting the third subplot
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

# Plotting the fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Adjusting the spacing between subplots
plt.tight_layout()

# Displaying the figure with subplots
plt.show()

```

```python
Ans 4 A bar plot, also known as a bar chart, is a type of visualization that uses rectangular bars to represent categorical data.
Each bar's length or height corresponds to the value or frequency of a particular category or group.
Bar plots are commonly used to compare and display data across different categories or groups.

Here are some key characteristics and reasons why bar plots are widely used:

Comparison of categorical data
Displaying frequency or count
Visualization of rankings or ordered data
Illustrating proportions or percentages
Communicating insights and findings

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.title("Company Profit")
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.title("Company Profit")
plt.show()
```

```python
Ans 5 A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a concise summary of a dataset's distribution.
It presents key statistical measures and provides insights into the variability, skewness, and potential outliers within the data.

Here are some key characteristics and reasons why box plots are commonly used:
Summary of distribution
Identification of outliers
Comparison of multiple group
Visualizing skewness and symmetry

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
